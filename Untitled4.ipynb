```python
import zipfile
import os

# Defina o caminho para o arquivo ZIP e o diretório de extração
zip_path = 'img.zip'
extract_path = '/content/descompactado/'
```

```python
os.makedirs(extract_path, exist_ok=True)
```

```python
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Arquivos descompactados com sucesso.")
```

```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Listar arquivos no diretório descompactado
files = os.listdir(extract_path)
png_files = [f for f in files if f.lower().endswith('.png')]

print(f"Número total de imagens PNG: {len(png_files)}")

# Mostrar algumas imagens para verificar
def mostrar_imagem(img_path):
    image = Image.open(img_path)
    plt.imshow(image)
    plt.axis('off')
    plt.show()

# Exibir as primeiras 5 imagens
for img_file in png_files[:5]:
    mostrar_imagem(os.path.join(extract_path, img_file))
```

```python
import os

# Listar todos os arquivos e subdiretórios no diretório de extração
for root, dirs, files in os.walk(extract_path):
    for file in files:
        print(os.path.join(root, file))
```
