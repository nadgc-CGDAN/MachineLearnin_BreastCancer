```python
import os

def mostrar_imagens():
    fig = plt.figure(figsize=(20, 20))

    for idx, image_path in enumerate(image_files):
        prediction_path = 'predictions.png'

        # Verifica se a imagem de predição foi gerada
        if os.path.exists(prediction_path):
            img = cv2.imread(prediction_path)

            if img is not None and img.size > 0:  # Verifica se a imagem não está vazia
                img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                ax = fig.add_subplot(2, 5, idx + 1)
                ax.imshow(img_rgb)
                ax.axis('off')
            else:
                print(f"Imagem {prediction_path} está vazia ou não pôde ser lida.")
        else:
            print(f"Arquivo {prediction_path} não encontrado.")

    plt.show()

mostrar_imagens()
```

```python
import cv2
import matplotlib.pyplot as plt
import os

def mostrar_imagens_detectadas():
    fig = plt.figure(figsize=(20, 10))  # Ajuste o tamanho da figura conforme necessário

    # Caminhos das imagens processadas com as detecções
    image_files = [
        "result_image1.jpg", "result_image2.jpg", "result_image3.jpg", "result_image25.jpg",
        "result_image42.jpg", "result_image45.jpg", "result_image59.jpg", "result_image90.jpg",
        "result_image99.jpg", "result_image9.jpg"
    ]

    for idx, image_path in enumerate(image_files):
        if os.path.exists(image_path):
            img = cv2.imread(image_path)

            if img is not None and img.size > 0:
                img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                ax = fig.add_subplot(2, 5, idx + 1)  # 2x5 (2 linhas e 5 colunas)
                ax.imshow(img_rgb)
                ax.axis('off')  # Desligar os eixos para uma visualização melhor
            else:
                print(f"Imagem {image_path} está vazia ou não pôde ser lida.")
        else:
            print(f"Arquivo {image_path} não encontrado.")

    plt.show()

mostrar_imagens_detectadas()
```

```python
import cv2
import matplotlib.pyplot as plt
def mostrar (frame):
  imagem = cv2.imread(frame)
  fig = plt.gcf() #função que chama uma figura do matplotlib
  fig.set_size_inches(18,10) #tamanho que vai ser gerada a saída
  plt.axis('off') #retira do formato padrão do matplotlib (como um plano cartesiano)
  plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
  plt.show()

mostrar('predictions.jpg')
```
